```python
# T: O(n+m), S: O(n+m)
from typing import Optional, List
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def treeQueries(self, root: Optional[TreeNode], queries: List[int]) -> List[int]:
        def height(root):
            if not root:    return 0
            lh = height(root.left)
            rh = height(root.right)
            l[root.val] = lh
            r[root.val] = rh
            return max(lh,rh) + 1
        def dfs(root, unaffected_height, depth):
            if not root:    return
            h[root.val] = unaffected_height
            dfs(root.left, max(unaffected_height, r[root.val]+depth), depth+1)
            dfs(root.right, max(unaffected_height, l[root.val]+depth), depth+1)

        l, r, h = {}, {}, {}
        height(root)
        dfs(root, 0, 0)
        return [h[q] for q in queries]
```

```python
# T: O(n+m), S: O(n+m)
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def treeQueries(self, root: Optional[TreeNode], queries: List[int]) -> List[int]:
        def height(root):
            # return tree hight at root
            if not root:    return 0
            lh = height(root.left)
            rh = height(root.right)
            l[root.val] = lh
            r[root.val] = rh
            return max(lh,rh) + 1
        def solve(root, maxa, depth):
            '''
            maxa: unaffected tree height
            depth: depth of node root
            '''
            if not root:    return
            h[root.val] = maxa
            solve(root.left, max(maxa, depth+r[root.val]), depth+1)
            solve(root.right, max(maxa, depth+l[root.val]), depth+1)
        l,r,h = {}, {}, {}
        '''
        l: node val to its left subtree height
        r: node val to its right subtree height
        h: node val to the delete this node, what's the new entire tree height
        '''
        height(root)
        solve(root.left, r[root.val], 1)
        solve(root.right, l[root.val], 1)
        return [h[q] for q in queries]

```
