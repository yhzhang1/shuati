```python
from typing import List
def combination(nums: List[int], d, n, s, cur, ans):
    if d == n:
        ans.append(cur[:])
        return
        
    for i in range(s, len(nums)):
        cur.append(nums[i])
        combination(nums, d+1, n, i+1, cur, ans)
        cur.pop()
```

```python
ans = list()
combination([1,2,2], 0, 2, 0, [], ans)
ans
```
