```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        counter2 = Counter(t)
        counter1 = Counter()
        matchCount = 0
        res = ''
        
        i = 0
        for j in range(len(s)):
            if s[j] in counter2:
                counter1[s[j]] += 1
                if counter1[s[j]] <= counter2[s[j]]:
                    matchCount += 1
                
            while i <= j and matchCount == len(t):
                if res == '' or j-i+1 < len(res):
                    res = s[i:j+1]
                if s[i] in counter1:
                    counter1[s[i]] -= 1
                    if counter1[s[i]] < counter2[s[i]]:
                        matchCount -= 1
                i += 1
                
        return res
                
                
            
```
