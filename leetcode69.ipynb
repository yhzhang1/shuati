```python
class Solution:
    def mySqrt(self, x: int) -> int:

        l, r = 0, x
        while l <= r:
            mid = (l+r) // 2
            print('l:{},r:{},m:{}'.format(l,r,mid))
            if mid**2 < x:
                l = mid + 1
            elif mid**2 == x:
                return mid
            else:
                r = mid - 1
                
        print(l, r)
                
        return r
```
