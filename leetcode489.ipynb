```python
# T: O(mn), S: O(mn)
# """
# This is the robot's control interface.
# You should not implement it, or speculate about its implementation
# """
#class Robot:
#    def move(self):
#        """
#        Returns true if the cell in front is open and robot moves into the cell.
#        Returns false if the cell in front is blocked and robot stays in the current cell.
#        :rtype bool
#        """
#
#    def turnLeft(self):
#        """
#        Robot will stay in the same cell after calling turnLeft/turnRight.
#        Each turn will be 90 degrees.
#        :rtype void
#        """
#
#    def turnRight(self):
#        """
#        Robot will stay in the same cell after calling turnLeft/turnRight.
#        Each turn will be 90 degrees.
#        :rtype void
#        """
#
#    def clean(self):
#        """
#        Clean the current cell.
#        :rtype void
#        """

class Solution:
    def cleanRoom(self, robot):
        """
        :type robot: Robot
        :rtype: None
        """
        def dfs(x, y, d):
            if (x,y) in visited:
                return
            visited.add((x,y))
            robot.clean()
            for _ in range(4):
                new_x = x + directions[d][0]
                new_y = y + directions[d][1]
                if robot.move():
                    dfs(new_x, new_y, d)
                    robot.turnRight()
                    robot.turnRight()
                    robot.move()
                    robot.turnRight()
                    robot.turnRight()
                d = (d+1) % 4
                robot.turnRight()

        directions = [(0,1), (1,0),(0,-1),(-1,0)]
        visited = set()
        dfs(0, 0, 0)
```

```python
# T: O(mn), S: O(mn)
# """
# This is the robot's control interface.
# You should not implement it, or speculate about its implementation
# """
#class Robot:
#    def move(self):
#        """
#        Returns true if the cell in front is open and robot moves into the cell.
#        Returns false if the cell in front is blocked and robot stays in the current cell.
#        :rtype bool
#        """
#
#    def turnLeft(self):
#        """
#        Robot will stay in the same cell after calling turnLeft/turnRight.
#        Each turn will be 90 degrees.
#        :rtype void
#        """
#
#    def turnRight(self):
#        """
#        Robot will stay in the same cell after calling turnLeft/turnRight.
#        Each turn will be 90 degrees.
#        :rtype void
#        """
#
#    def clean(self):
#        """
#        Clean the current cell.
#        :rtype void
#        """

class Solution:
    def move_to_direction(self, to_d, robot):
        if to_d == 'north':
            if self.d == 'east':    robot.turnLeft()
            elif self.d == 'west':  robot.turnRight()
            elif self.d == 'south': 
                robot.turnRight()
                robot.turnRight()
            self.d = 'north'
        elif to_d == 'south':
            if self.d == 'east':    robot.turnRight()
            elif self.d == 'west':  robot.turnLeft()
            elif self.d == 'north':
                robot.turnLeft()
                robot.turnLeft()
            self.d = 'south'
        elif to_d == 'east':
            if self.d == 'north':   robot.turnRight()
            elif self.d == 'south': robot.turnLeft()
            elif self.d == 'west':
                robot.turnLeft()
                robot.turnLeft()
            self.d = 'east'
        elif to_d == 'west':
            if self.d == 'north':   robot.turnLeft()
            elif self.d == 'south': robot.turnRight()
            elif self.d == 'east':
                robot.turnRight()
                robot.turnRight()
            self.d = 'west'

    def dfs(self, pos, visited, robot):
        print('pos:', pos)
        if pos in visited:
            return
        visited.add(pos)
        robot.clean()
        for next_d in ['north','south','east','west']:
            if next_d == 'north':   next_pos = (pos[0], pos[1]+1)
            elif next_d == 'south': next_pos = (pos[0], pos[1]-1)
            elif next_d == 'east':  next_pos = (pos[0]+1, pos[1])
            elif next_d == 'west':  next_pos = (pos[0]-1, pos[1])
            self.move_to_direction(next_d, robot)
            if robot.move():
                self.dfs(next_pos, visited, robot)
                self.move_to_direction(self.reverse_d[next_d], robot)
                robot.move()
                


    def cleanRoom(self, robot):
        """
        :type robot: Robot
        :rtype: None
        """
        self.reverse_d = {'north':'south','south':'north','east':'west','west':'east'}
        visited = set()
        self.d = 'north'
        self.dfs((0,0), visited, robot)
```
