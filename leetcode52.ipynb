```python
class Solution:
    def totalNQueens(self, n: int) -> int:
        self.col = [False for _ in range(n)]
        self.diag1 = [False for _ in range(2*n-1)]
        self.diag2 = [False for _ in range(2*n-1)]
        
        ans = self.dfs(n, 0)
        return ans
        
    def dfs(self, n, x) -> int:
        if x == n:
            return 1
            
        ans = 0
        for y in range(n):
            if not self.available(n, x, y):
                continue
            
            self.updateBoard(n, x, y, True)
            ans += self.dfs(n, x+1)
            self.updateBoard(n, x, y, False)
            
        return ans
            
    def available(self, n, x, y):
        return not self.col[y] and not self.diag1[x+y] and not self.diag2[x-y+n-1]
    
    def updateBoard(self, n, x, y, put):
        self.col[y] = put
        self.diag1[x+y] = put
        self.diag2[x-y+n-1] = put
        
        
            
    col = list()
    diag1 = list()
    diag2 = list()
            
    
```
