```python
class Solution:
    def reverse(self, x: int) -> int:
        ans = 0
        negative = x < 0
        x = abs(x)
        while x:
            digit = x % 10
            ans = 10*ans + digit
            x //= 10
            
        ans = -ans if negative else ans

        return ans if (-2**31 <= ans <= 2**31-1) else 0
```
