```python
# T:O(2^n), S:O(2^n)
from typing import List
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        def dfs(l, r, cur):
            if l > r:
                return
            if l == 0 and r == 0:
                ans.append(cur)
                return
                
            if l > 0:
                dfs(l-1, r, cur+'(')
            if r > 0:
                dfs(l, r-1, cur+')')
                
        ans = []
        dfs(n, n, '')
        return ans
```

```python
from typing import List

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        def bfs(l, r, cur, ans):
            if r < l:
                return
            if l == 0 and r == 0:
                ans.append(cur)
            
            if l > 0:
                bfs(l-1, r, cur+'(', ans)
            if r > 0:
                bfs(l, r-1, cur+')', ans)
                
        ans = list()
        bfs(n, n, '', ans)
        return ans
```

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        def dfs(l,r,cur):
            if l == n and r == n:
                ans.append(''.join(cur))
                return
            
            if l < n:
                cur.append('(')
                dfs(l+1,r,cur)
                cur.pop()
            if l > r and r < n:
                cur.append(')')
                dfs(l,r+1,cur)
                cur.pop()
                
        ans = []
        dfs(0,0,[])
        return ans
```
