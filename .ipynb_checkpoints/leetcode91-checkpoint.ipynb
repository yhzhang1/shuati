```python
def numDecodings(s: str) -> int:
    d = dict()

    for window in range(1, len(s)+1):
        start_idx = 0
        while start_idx+window <= len(s):
            substring = s[start_idx:start_idx+window] 
            if len(substring) == 1:
                if substring == '0':
                    d[substring] = 0
                else:
                    d[substring] = 1
            elif len(substring) == 2:
                if substring[1] == '0':
                    d[substring] = 0
                else:
                    d[substring] = 1
                if substring <= '26':
                    d[substring] += 1
            else:
                d[substring] = d[substring[1:]] + d[substring[2:]]
                
            start_idx += 1
            print(d)

    return d[s]
```
