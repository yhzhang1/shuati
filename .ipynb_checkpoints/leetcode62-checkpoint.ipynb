```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        dp = [[1]*n for _ in range(m)]
        
        for i in range(1,m):
            for j in range(1,n):
                dp[i][j] = dp[i][j-1] + dp[i-1][j]
                    
        return dp[-1][-1]
```

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        d = [[1] * m for _ in range(n)]
        
        for i in range(1, n):
            for j in range(1, m):
                d[i][j] = d[i-1][j] + d[i][j-1]
                
        return d[-1][-1]
```
