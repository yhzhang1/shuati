```python
from typing import List
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        def dfs(n, d, cur, ans):
            if d == n:
                ans.append(cur[:])
                return
            
            for i in range(n):
                flag = False
                if i in cur:
                # same column
                    continue
                for j in range(len(cur)):
                    if abs(j-len(cur)) == abs(cur[j]-i):
                    # diagonal
                        flag = True
                if flag:
                    continue
                        
                cur.append(i)
                dfs(n, d+1, cur, ans)
                cur.pop()
                
        ans = list()
        dfs(n, 0, list(), ans)
        ans_ = [[['.' for _ in range(n)] for _ in range(n)] for _ in range(len(ans))]

        for i in range(len(ans)):
            for j in range(n):
                ans_[i][j][ans[i][j]] = 'Q'
                
        for i in range(len(ans_)):
            for j in range(n):
                ans_[i][j] = ''.join(ans_[i][j])
                
        return ans_
```

```python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        self.board = ['.'*n for _ in range(n)]
        self.cols = [False for _ in range(n)]
        self.diag1 = [False for _ in range(2*n-1)]
        self.diag2 = [False for _ in range(2*n-1)]
        

        ans = list()
        self.dfs(n, 0, ans)
        return ans
        
    def dfs(self, n, x, ans):
        if x == n:
            ans.append(self.board[:])
            return

        for y in range(n):
            if not self.available(n, x, y):
                continue
                

            self.updateBoard(n, x, y, True)
            self.dfs(n, x+1, ans)
            self.updateBoard(n, x, y, False)
            
            
    def updateBoard(self, n, x, y, put):
        self.cols[y] = put
        self.diag1[x+y] = put
        self.diag2[x-y+n-1] = put
        
        if put:
            row = ''
            for i in range(n):
                if i == y:
                    row += 'Q'
                else:
                    row += '.'

            self.board[x] = row
        else:
            self.board[x] = ['.' for _ in range(n)]
                
                    
        
    def available(self, n, x, y):
        return not self.cols[y] and not self.diag1[x+y] and not self.diag2[x-y+n-1]
                
    board = list()
    cols = list()
    diag1 = list()
    diag2 = list()
        
        
```
