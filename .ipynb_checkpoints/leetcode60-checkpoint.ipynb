```python
class Solution:
    def getPermutation(self, n: int, k: int) -> str:
        ans = ''
        seq = [i+1 for i in range(n)]
        factorial = [0 for i in range(n)]
        factorial[0] = 1
        for i in range(1, n):
            factorial[i] = i*factorial[i-1]
        
        k -= 1
        
        for i in range(n):
            ans += str(seq.pop(k//factorial[n-1-i]))
            k %= factorial[n-1-i]
            
        return ans
```
