```python
class Solution:
    def divide(self, dividend: int, divisor: int) -> int:
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31
        
        
        if dividend == INT_MIN and divisor == -1:
            return INT_MAX
        
        ans = 0
        sign = 1
        if (dividend > 0 and divisor < 0) or (dividend < 0 and divisor > 0):
            sign = -1
        
        divd = abs(dividend)
        divs = abs(divisor)
        
        while divd >= divs:
            shift = 0
            while divd >= divs << shift:
                shift += 1
                
            ans += 1 << (shift-1)
            divd -= (divs << (shift-1))
            
        return ans if sign > 0 else -ans
```
